# Semana 1 - Python idiomatico (de Java a Python)

Objetivo: dejar de escribir Java en Python. No es aprender sintaxis desde cero, es interiorizar los patrones que usa Python de verdad y que vas a ver en todo el stack (pandas, PySpark, agregaciones).

Resuelve cada ejercicio en su celda de codigo. No hay soluciones puestas a proposito: el codigo lo escribes tu.


## 1. Comprehensions (list, dict, set)

- Dada una lista de numeros, crea una nueva lista con los cuadrados de los pares (en una sola linea).
- A partir de una lista de palabras, construye un dict {palabra: longitud}.
- Construye un set con las primeras letras de una lista de nombres.
- Reescribe un bucle for clasico con append como comprehension.


## 2. Unpacking y argumentos (*args, **kwargs)

- Intercambia dos variables sin variable temporal.
- Desempaqueta el primero y el resto de una lista: primero, *resto = ...
- Escribe una funcion que acepte numero variable de argumentos y devuelva su suma.
- Escribe una funcion que acepte **kwargs y los imprima como 'clave=valor'.


## 3. Context managers (with) y ficheros

- Abre un fichero de texto con 'with' y cuenta sus lineas.
- Escribe una lista de strings a un fichero, una por linea, usando 'with'.
- Explica en un comentario por que 'with' es mejor que abrir y cerrar a mano (el equivalente al try-with-resources de Java).


## 4. Type hints y dataclasses (vs POJOs de Java)

- Anota una funcion con tipos en parametros y retorno.
- Crea una dataclass Usuario con campos nombre: str y edad: int.
- Compara mentalmente con un POJO de Java: cuantas lineas te ahorras (sin getters, setters, equals, toString).


## 5. Generadores e iteradores (yield)

- Escribe un generador que produzca los numeros pares hasta n con 'yield'.
- Recorre el generador con un for y suma los valores.
- Explica en un comentario la diferencia entre devolver una lista (eager) y un generador (lazy), y por que importa con datasets grandes.


## 6. Idioms del dia a dia

- Recorre una lista con indice usando enumerate (no range(len(...))).
- Combina dos listas en pares con zip.
- Formatea un mensaje con f-strings.
- Maneja una conversion int() que puede fallar con try/except (estilo EAFP: pide perdon, no permiso).
